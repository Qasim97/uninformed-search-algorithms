```
# Muhammad Qasim Jamal
# 19I-0459
# CS-F

import math
import re
import sys
from random import sample


# function for file reading and data cleaning
def file_reading(file_path):
    graph_data_dict = {}
    data_dict_keys = []
    file = open(file_path, 'r')
    for line_data in file:
        line_data = line_data.replace('\n', '', 1)  # removing \n at the end f line
        parsed_line = re.split(r',{|},', line_data)  # splitting contents of line on the basis of comma & curly brackets
        # print(line_data)

        if int(parsed_line[0]) not in graph_data_dict.keys():  # check whether new child has been placed in dict or not
            child_num = int(parsed_line[0])
            data_dict_keys.append(child_num)
            graph_data_dict[child_num] = {}

        cost_key_parent = parsed_line[1]
        if parsed_line[1] == "":  # replacing the blank with None - related no parent
            cost_key_parent = None

        # adding to dictionary
        temp_dict = graph_data_dict[child_num]
        temp_dict[cost_key_parent] = float(parsed_line[2])
        # print(temp_dict)

    # print(data_dict_keys)
    # print(graph_data_dict)
    return graph_data_dict


data_dict = file_reading("./data0.txt")


# alternative dataset
# file_reading("./data1.txt")


def generate_possible_parent_nodes(node_seq):
    parent_seq = []
    for i in range(0, len(node_seq)):
        parent_seq.append(node_seq[i])
        for j in range(i + 1, len(node_seq)):
            parent_seq.append((node_seq[i] + node_seq[j]))

    # print(parent_seq)
    return parent_seq


def cost_calculation(input_order, file_dict):
    path_cost = 0
    minimum_cost = sys.maxsize
    visited_nodes = []
    found_flag = True
    for child in input_order:
        current_dict = file_dict[int(child)]
        if not visited_nodes:
            # print(current_dict)
            path_cost += current_dict[None]
        else:
            possible_parent_nodes = generate_possible_parent_nodes(visited_nodes)
            # print(possible_parent_nodes)
            for each_node in possible_parent_nodes:
                if len(each_node) > 1:
                    each_node = ','.join(each_node[i:i + 1] for i in range(0, len(each_node), 1))
                # print(each_node)
                if each_node in current_dict.keys():
                    if current_dict[each_node] < minimum_cost:
                        found_flag = False
                        minimum_cost = current_dict[each_node]

                if found_flag is True:
                    minimum_cost = current_dict[None]
            # print(minimum_cost)
            path_cost += minimum_cost

        visited_nodes.append(child)
        minimum_cost = sys.maxsize

    # print(path_cost)
    return path_cost


class Tree:
    def __init__(self, value):
        self.left = None
        self.right = None
        self.value = value
        # print(self.value)

    def insert(self, value):
        # Compare the new value with the parent node
        if self.value:
            if value < self.value:
                if self.left is None:
                    self.left = Tree(value)
                else:
                    self.left.insert(value)
            elif value > self.value:
                if self.right is None:
                    self.right = Tree(value)
                else:
                    self.right.insert(value)
        else:
            self.value = value

    def bfs(self, start_node):
        shortest_path = {"Minimum Cost": sys.maxsize, "Order": []}
        nodes_will_visit = []
        # This will add starting node (root node) into the list
        if start_node:
            nodes_will_visit.append(start_node)

        # This loop will traverse thorough the list
        while nodes_will_visit:

            # Pop the first node and check the children of that node
            current = nodes_will_visit.pop(0)

            # Will calculate the cost of the specified paths
            cost = cost_calculation(current.value, data_dict)

            #These lines will update the path with minimum cost
            if cost < shortest_path["Minimum Cost"]:
                shortest_path["Minimum Cost"] = cost
                shortest_path["Order"] = current.value

            if current.left:
                nodes_will_visit.append(current.left)
            if current.right:
                nodes_will_visit.append(current.right)

        return shortest_path

    def dfs(self, start_node):
        shortest_path = {"Minimum Cost": sys.maxsize, "Order": []}
        nodes_will_visit = []
        # This will add starting node (root node) into the list
        if start_node:
            nodes_will_visit.append(start_node)

        # This loop will traverse thorough the list
        while nodes_will_visit:

            # Pop the first node and check the children of that node
            current = nodes_will_visit.pop()

            # Will calculate the cost of the specified paths
            cost = cost_calculation(current.value, data_dict)

            #These lines will update the path with minimum cost
            if cost < shortest_path["Minimum Cost"]:
                shortest_path["Minimum Cost"] = cost
                shortest_path["Order"] = current.value

            if current.left:
                nodes_will_visit.append(current.left)
            if current.right:
                nodes_will_visit.append(current.right)

        return shortest_path

    def ucs(self, start_node):
        return self.dfs(start_node)



# print(data_dict)
lst_order = []
combinations = []
print("Enter a sequence: ")
for i in range(0, len(data_dict.keys())):
    lst_order.append(input())

combinations.append(lst_order)

root_node = Tree(0)

while len(combinations) < math.factorial(len(lst_order)):
    temp = sample(lst_order, len(lst_order))
    if temp not in combinations:
        combinations.append(temp)
        root_node.insert(temp)

path_thorugh_bfs = root_node.bfs(root_node)
print(f'Path With Minimum Cost (BFS): {path_thorugh_bfs}')
path_thorugh_dfs = root_node.dfs(root_node)
print(f'Path With Minimum Cost (DFS): {path_thorugh_dfs}')
path_thorugh_ucs = root_node.ucs(root_node)
print(f'Path With Minimum Cost (UCS): {path_thorugh_ucs}')


# print(combinations)
# print(len(combinations))


# Tree(lst_order)

# cost_calculation(lst_order, data_dict)

# Reference Links
# https://www.tutorialspoint.com/python_data_structure/python_binary_tree.htm (For Tree Creation Code)
# Stackoverflow 
```
